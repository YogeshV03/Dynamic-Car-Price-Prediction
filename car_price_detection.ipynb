```python
kms_min = df['Kilometers_Driven'].min()
kms_max = df['Kilometers_Driven'].max()

def scaler(x, min_, max_):
    return (x-min_)/(max_-min_)
```

```python
df['Kilometers_Driven'] = df['Kilometers_Driven'].apply(lambda x: scaler(x, kms_min,kms_max))
```

```python
df.head()
```

```python
e_min = df['Engine'].min()
e_max = df['Engine'].max()

def scaler(x, min_, max_):
    return (x-min_)/(max_-min_)
```

```python
df['Engine'] = df['Engine'].apply(lambda x: scaler(x, e_min,e_max))
```

```python
m_min = df['Mileage'].min()
m_max = df['Mileage'].max()

def scaler(x, min_, max_):
    return (x-min_)/(max_-min_)
```

```python
df['Mileage'] = df['Mileage'].apply(lambda x: scaler(x, m_min,m_max))
```

```python
p_min = df['Power'].min()
p_max = df['Power'].max()

def scaler(x, min_, max_):
    return (x-min_)/(max_-min_)
```

```python
df['Power'] = df['Power'].apply(lambda x: scaler(x, p_min,p_max))
```

```python
df.head()
```

```python
y_min = df['Year'].min()
y_max = df['Year'].max()

def scaler(x, min_, max_):
    return (x-min_)/(max_-min_)
```

```python
df['Year'] = df['Year'].apply(lambda x: scaler(x, y_min,y_max))
```
